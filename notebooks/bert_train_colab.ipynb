## 3. Upload data (if not in repo)

If `data/support_tickets.json` is not in your repo (e.g. too large for GitHub), upload it here.

```python
import os

data_path = "data/support_tickets.json"

if not os.path.exists(data_path):
    print("Data file not found — upload it now:")
    from google.colab import files
    uploaded = files.upload()  # select support_tickets.json from your PC
    os.makedirs("data", exist_ok=True)
    for filename, content in uploaded.items():
        with open(data_path, "wb") as f:
            f.write(content)
    print(f"Saved to {data_path}")
else:
    print(f"Data file found at {data_path}")
```
